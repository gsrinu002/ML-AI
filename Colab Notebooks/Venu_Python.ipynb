```python
import tkinter as tk

# Create the main window
root = tk.Tk()
root.title("Tkinter Calculator")
root.geometry("300x400")

# Entry widget for input and result display
entry = tk.Entry(root, font=("Arial", 20), bd=10, relief=tk.RIDGE, justify="right")
entry.grid(row=0, column=0, columnspan=4, pady=10, padx=10)

# Function to handle button clicks
def on_click(value):
    if value == "=":
        try:
            result = str(eval(entry.get()))
            entry.delete(0, tk.END)
            entry.insert(tk.END, result)
        except:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    elif value == "C":
        entry.delete(0, tk.END)
    else:
        entry.insert(tk.END, value)

# Button labels
buttons = [
    ['7', '8', '9', '/'],
    ['4', '5', '6', '*'],
    ['1', '2', '3', '-'],
    ['C', '0', '=', '+']
]

# Create buttons dynamically using grid
for i, row in enumerate(buttons):
    for j, btn_text in enumerate(row):
        button = tk.Button(root, text=btn_text, font=("Arial", 18), padx=20, pady=20,
                           command=lambda val=btn_text: on_click(val))
        button.grid(row=i+1, column=j, sticky="nsew")

# Make buttons expand with window
for i in range(4):
    root.grid_columnconfigure(i, weight=1)
    root.grid_rowconfigure(i+1, weight=1)

# Run the application
root.mainloop()
```

```python
import json

# json.dumps() : converts dict -> jsonobject  & list -> jsonarray

emp_object_dic= {'emp_name':'venu','emp_salary':67000,'emp_id':1001}
print("emp object as dictionary..", emp_object_dic)
emp_obj_json=json.dumps(emp_object_dic)
print("emp object as json ..",emp_obj_json)

emp_list_python=[{'emp_name':'venu','emp_salary':67000,'emp_id':1001},{'emp_name':'srinivas','emp_salary':85000,'emp_id':1002}]
print(emp_list_python)
emp_list_json=json.dumps(emp_list_python)
print(emp_list_json)
```

```python
#json.loads() : converts json_obj-> python_dict json_array-> python_list

emp_obj_json='{"emp_name":"venu","emp_salary":67000,"emp_id":1001}'
print(emp_obj_json)

emp_obj_dict=json.loads(emp_obj_json)
print(emp_obj_dict)

print(emp_obj_dict['emp_name'])
print(emp_obj_dict['emp_salary'])
print(emp_obj_dict['emp_id'])

emp_list_json='[{"emp_name":"venu","emp_salary":67000,"emp_id":1001},{"emp_name":"srinivas","emp_salary":85000,"emp_id":1002}]'
print(emp_list_json)
emp_list_python=json.loads(emp_list_json)
print(emp_list_python)

print(emp_list_python[0]['emp_name'])
print(emp_list_python[0]['emp_salary'])
print(emp_list_python[0]['emp_id'])

print(emp_list_python[1]['emp_name'])
print(emp_list_python[1]['emp_salary'])
print(emp_list_python[1]['emp_id'])
```

```python
class Student:
  # constructor
  def __init__(self,name,age,marks):
    self.name=name
    self.age=age
    self.marks=marks
  # class method
  def display(self):
    print(self.name)
    print(self.age)
    print(self.marks)
```

```python
s1=Student("venu",27,90)
s1.display()

s2=Student("sirisha",25,95)
s2.display()
```

```python
class Bank:
  def __init__(self,deposit,withdraw):
    self.deposit=deposit
    self.withdraw=withdraw
  def display(self):
    print("deposit amount..",self.deposit)
    print("withdraw amount..",self.withdraw)
    avail_balance=self.deposit-self.withdraw
    print("available balance..",avail_balance)

axis=Bank(10000,2500)
axis.display()
```

```python
sbi=Bank(100000,40000)
sbi.display()
```
